## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.

- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

### Cleaning Data

#### Fixing Data Type, Categorical Label, and Data Efficiency Problem

```python
# Mengubah tipe data dteday menjadi datetime
day_df['dteday'] = pd.to_datetime(day_df['dteday'])

# Mapping nilai kategori agar lebih informatif
day_df['season'] = day_df['season'].map({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})
day_df['yr'] = day_df['yr'].map({0: '2011', 1: '2012'})
day_df['mnth'] = day_df['mnth'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['weathersit'] = day_df['weathersit'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Heavy Rain/Ice'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})

# Mengubah tipe data menjadi 'category' (Tambahan Penting)
categorical_columns = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
for col in categorical_columns:
    day_df[col] = day_df[col].astype('category')

# Menampilkan hasil pembersihan dan verifikasi tipe data
print(day_df.info())
day_df.head()
```

**Insight:** (Opsional)
- Konversi Tipe Data Sukses: Atribut 'dteday' sudah bertipe datetime64, yang memungkinkan saya melakukan analisis berbasis waktu secara akurat menggunakan fungsi-fungsi *time-series* di Python.
- Peningkatan Keterbacaann: Dengan melakukan *mapping* pada kolom kategorikal, data sekarang jauh lebih mudah dipahami. Contohnya, kita bisa langsung melihat bahwa tanggal 01-01-2011 adalah hari Sat (Sabtu) dengan cuaca Misti/Cloudy di musim Spring, bukan lagi sekadar angka kode.
- Efisiensi Memori: Penggunaan tipe data category berhasil menurunkan penggunaan memori dari 91.5 KB menjadi 58.0 KB (berkurang sekitar 36%). Hal ini bermanfaat untuk performa komputasi jika dataset bertambah besar.
- Kesiapan Visualisasi: Karena atribut kategorikal sudah memilik label string, maka saat saya membuat grafik nanti, sumbu grafik akan otomatis menampilkan nama musim/cuaca tanpa perlu kita atur secara menual satu per satu.

## Exploratory Data Analysis (EDA)

### Explore Pengaruh Cuaca Terhadap Penyewaan (2012)
Bagian ini menganalisis dampak kondisi cuaca terhadap jumlah penyewaan sepeda khusus pada tahun 2012. Analisis ini mencakup statistik deskriptif per kondisi cuaca serta melihat perbandingan perilaku antara pengguna tipe *casual* dan *registered*.

```python
# Mengambil data tahun 2012
df_2012 = day_df[day_df['yr'] == '2012']

# Agregasi penyewaan berdasarkan kondisi cuaca
weather_analysis = df_2012.groupby('weathersit', observed=True)['cnt'].agg(['mean', 'max', 'min', 'sum']).reset_index()

# Tambahan: Melihat persentase distribusi casual vs registered per cuaca
weather_user_type = df_2012.groupby('weathersit', observed=True)[['casual', 'registered']].mean().reset_index()

print("--- Analisis Cuaca 2012 ---")
print(weather_analysis)
print("\n--- Rata-rata Tipe Pengguna per Kondisi Cuaca ---")
print(weather_user_type)
```

- Dampak SIgnifikan Cuaca Buruk: Terjadi penurunan drastis pada rata-rata penyewaan saat cuaca berubah menjadi *Light Snow/Rain* hanya mencapai 2.126. Menunjukkan adanya risiko operasional yang tinggi dan penurunan permintaan hingga lebih dari 60% saat cuaca memburuk.

Kemungkinan besar pengguna 'registered' adalah pekerja/pelajar yang tetap harus bermobilitas meski cuaca kurang bersahabat, sementara pengguna 'casual' sangat dipengaruhi oleh kenyamanan cuaca.

### Explore Perbandingan Hari Kerja vs Hari Libur di Musim Gugur
Bagian ini mengeksplorasi perilaku penyewaan sepeda selama musim gugur (*Fall*) untuk membandingkan apakah terdapat perbedaan signifikan antara hari kerja (*workingday*) dan hari libur (*holiday/weekend*), serta melihat tren spesifiknya pada setiap bulan di musim tersebut.

```python
# Mengambil data musim gugur (Fall)
fall_df = day_df[day_df['season'] == 'Fall']

# Agregasi berdasarkan workingday untuk melihat rata-rata total penyewaan
workingday_analysis = fall_df.groupby('workingday', observed=True)['cnt'].agg(['mean', 'std', 'count']).reset_index()

# Tambahan: Melihat perbedaan performa pada setiap bulan di musim gugur
fall_monthly_analysis = fall_df.groupby(['mnth', 'workingday'], observed=True)['cnt'].mean().unstack()

print("--- Analisis Hari Kerja vs Libur (Musim Gugur) ---")
print(workingday_analysis)
print("\n--- Rata-rata Penyewaan per Bulan di Musim Gugur ---")
print(fall_monthly_analysis)
```

**Insight:** (Opsional)
- Korelasi Positif Kuat: Koefisien Pearson sebesar 0,6275 menunjukkan hubungan positif yang kuat antara suhu dan jumlah penyewaan. Secara statistik, kenaikan suhu cenderung diikuti dengan kenaikan jumlah penyewa.
- Ambang Batas Kenyamanan: Penyewaan mencapai puncaknya pada kategori suhu "Hangat" (5.547 sewa) dan "Panas" (5.610 sewa).
- Dampak Suhu Dingin: Sebaliknya, pada kategori suhu "Dingin", rata-rata penyewaan turun drastis hingga angka 1.819. Ini membuktikan bahwa kenyamanan termal merupakan faktor krusial yang menentukan keputusan masyarakat untuk menggunakan sepeda.

### Explore Tren Pertumbuhan Tahunan
Bagian ini untuk mengeksplorasi total penyewaa secara tahunan untuk melihat pertumbuhan bisnis dari tahun 2011 ke 2012.

```python
# Hitung total penyewaan harian
yearly_rentals = day_df.groupby('yr', observed=False)['cnt'].sum().reset_index()

# Cek apakah perlu di-mapping (Hanya jika nilainya masih 0 atau 1)
if yearly_rentals['yr'].iloc[0] in [0, 1]:
    yearly_rentals['yr'] = yearly_rentals['yr'].map({0: 2011, 1: 2012})

print("Total Penyewaan Tahunan:")
print(yearly_rentals)
```

**Insight:** (Opsional)
- Pertumbuhan Bisnis Masif: Terjadi lonjakan penyewaan yang sangat signifikan sebesar kurang lebih 64,8% secara *Year-over-Year*. Total sewa meningkat dari 1,24 juta pada tahun 2011 menjadi 2,04 juta pada tahun 2012.

### Explore Pengaruh Suhu terhadap Minat Sewa
Bagian ini melihat hubungan antara suhu rata-rata harian dengan jumlah total penyewaan sepeda.

```python
# Menghitung korelasi Pearson antara suhu (temp) dan jumlah sewa (cnt)
correlation = day_df['temp'].corr(day_df['cnt'])

# Mendapatkan statistik deskriptif untuk kedua variabel
temp_stats = day_df[['temp', 'cnt']].describe()

print(f"Koefisien Korelasi Pearson antara Suhu dan Jumlah Sewa: {correlation:.4f}")
print("\nStatistik Deskriptif Suhu vs Jumlah Sewa:")
display(temp_stats)

# Melihat rata-rata penyewaan dalam rentang suhu tertentu (Binning sederhana)
# Mengelompokkan suhu menjadi 4 bagian (Dingin, Sejuk, Hangat, Panas)
day_df['temp_bins'] = pd.cut(day_df['temp'], bins=4, labels=['Dingin', 'Sejuk', 'Hangat', 'Panas'])
temp_analysis = day_df.groupby('temp_bins', observed=False)['cnt'].mean().reset_index()

print("\nRata-rata Penyewaan berdasarkan Kategori Suhu:")
display(temp_analysis)
```

**Insight:** (Opsional)
- Performa Hari Kerja Lebih Unggul: Di musim gugur, rata-rata penyewaan pada hari kerja sebesar 5.718, sedikit lebih tinggi dibandingkan hari libur/akhir pekan yang sebesar 5.474.
- Konsistensi vs Variansi: Nilai standar deviasi pada hari libur adalah 1.433, sedikit lebih rendah daripada hari kerja yang mencapai 1.470. Namun, keduanya menunjukkan angka cukup tinggi, menandakan flukatasi penyewaan di musim gugur masih dipengaruhi oleh faktor harian lainnya (seperti suhu spesifik harian).
- Tren Bulanan yang Unik: Pada bulan Agustus, terdapat perbedaan kontras di mana hari kerja jauh lebih tinggi sebesar 5.908 dibandingkan hari libur yang hanya 4.963. Namun, pada bulan Juni dan September, terjadi anomali di mana hari libur justru memiliki rata-rata penyewaan jauh lebih tinggi daripada hari kerja. Hal ini menunjukkan bahwa puncak musim gugur atau cuaca yang sejuk, masyarakat lebih cenderung menggunakan sepeda untuk rekreasi/menimati hari libur di akhir pekan.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh perbedaan kondisi cuaca (weathersit) terhadap rata-rata jumlah penyewaan sepeda harian selama tahun 2012?

```python
plt.figure(figsize=(10, 6))

# Menggunakan sns.barplot dengan minimalisasi elemen hiasan
ax = sns.barplot(
    x='weathersit',
    y='mean',
    data=weather_analysis,
    palette='viridis'
)

# Menambah label nilai di atas batang untuk meningkatkan Data-Ink Ratio (informasi langsung)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points')

plt.title('Rata-rata Penyewaan Sepeda Berdasarkan Kondisi Cuaca (2012)', fontsize=14, pad=20)
plt.xlabel(None) # Menghapus label sumbu X karena kategori sudah jelas (meningkatkan ink ratio)
plt.ylabel('Rata-rata Jumlah Sewa', fontsize=12)

# Menerapkan sns.despine() untuk menghapus garis tepi atas dan kanan (Prinsip Tufte)
sns.despine()
plt.tight_layout()
plt.show()
```

- Bisa dipertimbangkan program loyalitas atau diskon khusus saat cuaca mendung/hujan ringan untuk menjaga kestabilan jumlah sewa.

- Kenaikan dari 1,24 juta ke 2,05 juta penyewaan menunjukkan bahwa layakan bike sharing ini telah melewati fase perkenalan yang kini berada di fase pertumbuhan cepat. Menunjukka indikator kuat bahwa pasar semakin menerima penggunaan sepeda sebagai moda transportasi atau rekreasi.

### Pertanyaan 4: Sejauh mana korelasi antara suhu udara (temperature) terhadap jumlah total penyewaan sepeda harian di seluruh musim?

```python
plt.figure(figsize=(10, 6))

# Menggunakan regplot dengan transparansi (alpha) untuk kejujuran sebaran data (Integritas)
sns.regplot(
    x='temp',
    y='cnt',
    data=day_df,
    scatter_kws={'alpha':0.3, 'color':'#2E86C1'},
    line_kws={'color':'#E74C3C', 'linewidth': 3},
    ci=None
)

# Menambahkan koefisien korelasi ke dalam plot (Integritas statistik)
corr_value = day_df['temp'].corr(day_df['cnt'])
plt.text(0.1, 8000, f'Korelasi Pearson: {corr_value:.2f}',
         fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.title('Korelasi Suhu (Normalized) terhadap Jumlah Penyewaan', fontsize=15, loc='left', pad=20)
plt.xlabel('Suhu Udara (Normalized)', fontsize=12)
plt.ylabel('Total Penyewaan Harian', fontsize=12)

sns.despine()
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

**Insight:** (Opsional)
- Terdapat hubungan linear positif yang kuat dengan korelasi pearson sebesar 0.63 anata suhu dan jumlah sewa sepeda.
- Semakin hangat suhu udara (hingga titik tertentu), semakin tinggi minat masyarakat untuk bersepeda.
- Sebaran data menunjukkan kepadatan yang tinggi pada suhu menengah ke atas, yang berarti cuaca hangat adalah pemicu utama pendapatan bisnis ini.

## Analisis Lanjutan (Opsional)

Pada bagian ini, saya menerapkan dua teknik analisis tambahan untuk mendapatkan insight yang lebih komprehensif:
1. RFM Analysis: Untuk melihat kapan transaksi terakhir terjadi (Recency), seberapa sering volume tinggi tercapai (Frequency), dan total volume penyewaan (Monetary) berdasarkan bulan.
2. Manual Grouping (Clustering): Mengategorikan hari-hari ke dalam tingkat performa penyewaan (Low, Mediun, High) untuk melihat distribusi produktivitas harian tanpa menggunakan algoritma Machine Learning.

### Analisis Lanjutan 1 (RFM)

```python
# Menentukan tanggal referensi (satu hari setelah data terakhir)
recent_date = day_df['dteday'].max() + pd.Timedelta(days=1)

# Recency: Selisih hari sejak transaksi terakhir di bulan tersebut
# Frequency: Jumlah hari transaksi dalam sebulan
# Monetary: Total volume penyewaan dalam sebulan
rfm_df = day_df.groupby(by="mnth", observed=False).agg({
    "dteday": lambda x: (recent_date - x.max()).days,
    "cnt": ["count", "sum"]
})

rfm_df.columns = ["recency", "frequency", "monetary"]

print("RFM Analysis Berdasarkan Bulan:")
display(rfm_df)
```

**Insight:**
- Nilai Recency menurun drastis hingga 1 hari pada bulan Desember. Menunjukkan bahwa data up-to-date hingga akhir tahun 2012, memberikan validitas tinggi untuk perencanaan strategi di awal tahun berikutnya.
- Jumlah hari transaksi (frequency) stabil di angka 60-62 hari (total dari 2011 & 2012). Mengindikasikan layanan beroperasi penuh tanpa gangguan teknis atau penutupan operasional yang signifikan di setiap bulannya.
- Terjadi tren kenaikan monetary yang konsisten mulai dari Maret dan mencapai puncaknya pada Agustus (351.194 penyewaan). Menariknya, performa mulai menurun saat memasuki November-Desember.
- Periode Mei hingga Oktober adalah "Golden Months" di mana volume penyewaan selalu di atas 300.000. Perusahaan harus memastikan armada sepeda dalam kondisi prima di bulan-bulan ini.

### Analisis Lanjutan 2 (Manual Grouping)

```python
# Menentukan kriteria binning (Low, Medium, High)
bins = [0, 3100, 6000, day_df['cnt'].max()]
labels = ['Low Performance', 'Medium Performance', 'High Performance']
day_df['rental_category'] = pd.cut(day_df['cnt'], bins=bins, labels=labels)

# Persiapan canvas
plt.figure(figsize=(10, 6))

# Menggunakan palet warna yang konsisten dan kontras
ax = sns.countplot(
    x='rental_category',
    data=day_df,
    palette='viridis',
    order=labels
)

# Menambahkan Anotasi Angka
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())} hari',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 10),
                textcoords='offset points',
                fontsize=11,
                fontweight='bold')

# Membersihkan Elemen Visual
plt.title('Distribusi Performa Penyewaan Harian (2011-2012)', fontsize=14, pad=20, loc='left')
plt.xlabel(None)
plt.ylabel('Jumlah Hari')
plt.ylim(0, day_df['rental_category'].value_counts().max() + 50) # Memberi ruang untuk anotasi

sns.despine() # Mengurangi Chartjunk (Prinsip Tufte)
plt.tight_layout()
plt.show()

# Menampilkan proporsi untuk mendukung Integritas Data
print("Proporsi Kategori Performa (Persentase):")
print(day_df['rental_category'].value_counts(normalize=True) * 100)
```

**Insight:**
- Lebih dari separuh total hari operasional (51% atau 373 hari) berada pada kategori Medium Performance (3.100 - 6.000 sewa). Ini menunjukkan bahwa bisnis memiliki basis permintaan harian yang stabil dan sehat.
- Terdapat proporsi yang hampir seimbang antara Low Performance (24,3%) dan High Performance (24,6%).
- Low Performance (178 hari): Kemungkinan besar terjadi pada hari-hari di musim dingin atau saat kondisi cuaca ekstrim (seperti yang terlihat pada analisis cuaca sebelumnya).
- High Performance (180 hari): Menunjukkan bahwa dalam 2 tahun, ada sekitar 6 bulan total di mana bisnis mencapai kapasitas maksimal (di atas 6.000 sewa).
- Grafik ini mengonfirmasi bahwa pertumbuhan yang ditemukan di analisis tahunan (64,8%) didorong oleh banyaknya hari yang berhasil masuk ke kategori Medium dan High, bukan sekadar lonjakan di hari-hari tertentu saja.
